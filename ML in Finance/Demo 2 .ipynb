## OLS REGRESSION

```python
T = 750 #Choose a sample size
mu = (0,0,0) #Generate sample for Y and X
cov = [[1,0,0],[0,1,0],[0,0,1]]

F = np.random.multivariate_normal(mu,cov,T)
X = np.column_stack((np.ones((T,1)),F))
N = X.shape


beta = np.array([0.56,2.05,1,78])
beta.shape = (N[1],1)

#GEnerate sampel fo Y
Y = np.matmul(X,beta) + np.random.normal(0,1,(T,1))
```

```python
invXX = np.linalg.inv(X.transpose()@X)

beta_hat = invXX@X.transpose()@Y

y_hat = X@beta_hat

residuals = Y - y_hat
```

```python
sigma2 = (1/T)*residuals.transpose()@residuals

sigma = np.sqrt(sigma2)

obj_func = sigma2 + 0.5*abs(beta_hat) + 0.5*(beta_hat)**2

varcov_beta_heat = (sigma2)*invXX

std_beta_hat = np.sqrt(T*np.diag(varcov_beta_heat))

R_squared = 1- (residuals.transpose()@residuals)/(T*np.)

adj_R_squared = 1- (1-R_squared)*(T-1)(T-N[1])

```
